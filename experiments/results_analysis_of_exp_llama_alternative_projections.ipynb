```python
api = wandb.Api(timeout=120)

prj = "gmum/bloraxs3"
runs = api.runs(prj,)
```

```python
def extract_wandb_params(config_dict):
    """
    Extracts task, reconstruction type, seed, and start time from W&B config.
    """
    def get_val(key):
        return config_dict.get(key, {}).get('value')

    # 1. Extract basic params
    extracted = {
        "task_name": get_val("experiment/task") or get_val("task"),
        "reconstruction_type": get_val("reconstruction_type"),
        "seed": get_val("experiment/seed") or get_val("seed"),
        "lora_r": get_val("experiment/lora_r") or get_val("lora_r"),
        "start_time": None
    }

    # 2. Extract Date/Time from the nested _wandb metadata
    # Path: _wandb -> value -> e -> [dynamic_hash_key] -> startedAt
    wandb_metadata = config_dict.get("_wandb", {}).get("value", {}).get("e", {})
    
    for run_id in wandb_metadata:
        start_str = wandb_metadata[run_id].get("startedAt")
        if start_str:
            # Returns the ISO string: "2026-01-15T12:09:44.079821Z"
            extracted["start_time"] = start_str
            break

    return extracted
```

```python
def extract_stats(df, dirs = ["map_metrics/", "laplace_LORAXS_KRON/", "laplace_LORAXS_DIAG/", "laplace/base_"]):
    ds = {}

    for dir in dirs:
        d = df[~df[f'{dir}epoch'].isna()]
        d = d[[c for c in d.columns if dir in c]]

        ds[dir] = d

    return ds

def load_and_extract(api, run_name):
    run = api.run(run_name)
    df = run.history(samples=100000)
    # print([f for f in list(df.columns) if "checkpoint" not in f])
    ds = extract_stats(df)
    return ds
```

```python
def extract_metrics_at_epoch(
    dataframe,
    epoch,
    wandb_dirs=[
        "map_metrics/",
        "laplace_LORAXS_KRON/",
        "laplace_LORAXS_DIAG/",
        "laplace/base_",
    ],
):

    metrics_selected = {}
    for prefix_dir in wandb_dirs:
        try:
            dir_results = dataframe[prefix_dir]
            dir_metrics = dir_results[dir_results[prefix_dir + "epoch"] == epoch]  # Filter by epoch
            dir_metrics = dir_metrics.to_dict(orient='records')[0]

            metrics_selected.update(dir_metrics)
        except Exception as e:
            print(f"Error processing {prefix_dir}: {e}")

    return metrics_selected
```

```python
KEEP_RUNS_WITH_PATTERNS = ["20260319"]
CHECKPOINT_EPOCH = 5

RELOAD = False
```

```python
if RELOAD:
    cfgs = []
    for run in runs:
        cfg = extract_wandb_params(json.loads(run.json_config))

        if not any([(pref in run.name) for pref in KEEP_RUNS_WITH_PATTERNS]):
            print(f"Skipping run: {run.name}")
            continue

        print(f"Parsing run: {run.name}")

        cfg["name"] = run.name
        cfg["state"] = run.state
        cfg["id"] = prj+"/"+run.id

        try:
            stats = load_and_extract(api, prj+"/"+run.id)
            metrics_at_epoch = extract_metrics_at_epoch(stats, epoch=CHECKPOINT_EPOCH)
            cfg.update(metrics_at_epoch)
        except Exception as e:
            print(f"  Failed to load stats from {run.name}", e)
            continue

        cfgs.append(cfg)
    
```

```python
try:
    df0 = pd.DataFrame(cfgs)
    df0.to_csv(f"results/results_analysis_of_exp_llama_alternative_projections_{CHECKPOINT_EPOCH}.csv", index=False)
except Exception as e:
    print("Failed to save results to CSV:", e)
```

```python

def create_metric_plots_single_row(df):
    """
    Alternative layout: One row per dataset, with all 4 combinations as different lines on the same plot.
    """
    
    # Define the order for projections
    projection_order = ['cca', 'svd', 'dct', 'random'] # 'cca-1/2_svd', 'dct-1/2_svd'
    projection2name = {
        'cca': 'W-SVD',
        'cca-1/2_svd': 'CCA-1/2-SVD',
        'svd': 'SVD',
        'dct-1/2_svd': 'DCT-1/2-S',
        'dct': 'DCT',
        'random': 'RAND'
    }
    
    # Define datasets
    datasets = ['obqa', 'arc-e']#, 'arc-c']
    
    # Define posteriors and lora_r values
    posteriors = ['KRON', 'DIAG']
    lora_r_values = [32, 64]
    
    # Create figure with 4 rows (one per dataset)
    fig, axes = plt.subplots(2, 3, figsize=(7.5, 1.5*2.0+0.25))
    # fig.suptitle(f'{metric} by Projection Method', fontsize=16, y=0.995)
    
    # Color and style map for different combinations
    styles = {
        ('KRON', 32): {'color': 'dodgerblue', 'linestyle': '-', 'marker': '^'},
        ('KRON', 64): {'color': 'blue', 'linestyle': '-', 'marker': 's'},
        ('DIAG', 32): {'color': 'salmon', 'linestyle': '--', 'marker': '^'},
        ('DIAG', 64): {'color': '#d62728', 'linestyle': '--', 'marker': 's'}
    }
    
    for col_ix, metric in enumerate(["Accuracy", "ECE", "NLL",]):

        # Iterate through datasets
        for row_idx, dataset in enumerate(datasets):
            df_dataset = df[df['dataset'] == dataset]
            
            ax = axes[row_idx, col_ix]

            def draw(ax):
                # Plot all combinations on the same plot
                for posterior in posteriors:
                    for lora_r in lora_r_values:
                        # Filter data for this combination
                        df_filtered = df_dataset[
                            (df_dataset['posterior'] == posterior) & 
                            (df_dataset['lora_r'] == lora_r)
                        ]

                        # print(dataset, posterior, lora_r, len(df_filtered))

                        
                        # Calculate median values for each projection
                        median_values = []
                        x_positions = []
                        
                        for proj_idx, projection in enumerate(projection_order):
                            df_proj = df_filtered[df_filtered['projection'] == projection]
                            if len(df_proj) > 0:
                                # median_val = df_proj[metric].mean()
                                median_val = df_proj[metric].median()
                                median_values.append(median_val)
                                x_positions.append(proj_idx)
                        
                        # Plot the line
                        if len(median_values) > 0:
                            style = styles[(posterior, lora_r)]
                            ax.plot(x_positions, median_values, 
                                label=f'{posterior[:1]}, r={lora_r}',
                                linewidth=1, 
                                markersize=4,
                                **style)

            draw(ax)
                
            if row_idx==len(datasets)-1:
                # Set x-axis labels
                ax.set_xticks(range(len(projection_order)))
                ax.set_xticklabels([projection2name.get(l,l) for l in projection_order], rotation=45, ha='right')
            else:
                ax.set_xticks([])
            
            # Set labels
            metric2label = {"Accuracy": "$\\uparrow$Accuracy", "ECE": "$\\downarrow$ECE", "NLL": "$\\downarrow$NLL"}
            ax.set_ylabel(metric2label[metric], fontsize=11)
            if col_ix==1:
                ax.set_title(dataset.upper(), fontsize=12)#, fontweight='bold', pad=10)
            
            if row_idx==0 and col_ix==2:
                # Add legend
                ax.legend( framealpha=0.7,
                        #   bbox_to_anchor=(1, 1), 
                        loc='lower right',
                          fontsize=7)
            
            # Add grid
            ax.grid(False)#, alpha=0.3, linestyle='--')
            
            # Format y-axis    
            ax.ticklabel_format(style='plain', axis='y')

            # ax.set_yscale('log')

            if col_ix==0 or col_ix==2:

                # [x_start, y_start, width, height] in relative coordinates (0 to 1)
                if col_ix==0:
                    ax_ins = ax.inset_axes([0.025, 0.025, 0.3, 0.55]) 
                else:
                    ax_ins = ax.inset_axes([0.025, 0.3-0.025, 0.3, 0.55])

                # 3. Plot the same data (or a subset) on the inset
                # ax_ins.plot(x, y, color='indianred')
                draw(ax_ins)

                # 4. Define the "Zoom" area
                    
                # 2. Create the inset axes
                if col_ix==0 and row_idx==0:
                    x1, x2, y1, y2 = -0.1, 1.1, 0.755, 0.805 # Focus area
                elif col_ix==0 and row_idx==1:
                    x1, x2, y1, y2 = -0.1, 1.1, 0.8125, 0.8425 # Focus area
                elif col_ix==2 and row_idx==0:
                    x1, x2, y1, y2 = -0.1, 1.1, 0.6225, 0.71 # Focus area
                elif col_ix==2 and row_idx==1:
                    x1, x2, y1, y2 = -0.1, 1.1, 0.4525, 0.585 # Focus area
                else:
                    raise Exception("Unexpected combination of col_ix and row_idx")

                ax_ins.set_xlim(x1, x2)
                ax_ins.set_ylim(y1, y2)

                # 5. Add custom ticks for the inset if desired
                # ax_ins.set_xticks([2.5, 3.0, 3.5])
                # ax_ins.set_yticks([1.2, 1.5, 1.8])
                ax_ins.set_xticks([])
                ax_ins.tick_params(axis='y', which='major', labelsize=10)
                ax_ins.yaxis.tick_right()

                # ax_ins.grid(True, alpha=0.9, linestyle='-')

                ax_ins.set_title("zoom", fontsize=9, pad=2)
                # 6. Draw lines connecting the zoom area to the inset
                # ax.indicate_inset_zoom(ax_ins, edgecolor="black")

        
        # Set common x-label
        if col_ix==1:
            ax.set_xlabel('Projection', fontsize=12)#, fontweight='bold')


    
    # Adjust layout
    plt.tight_layout()
    
    # Save the figure
    # plt.savefig(output_path, dpi=300, bbox_inches='tight')
    # print(f"Plot saved to {output_path}")
    
    return fig
```

```python
def stat(series):
    if pd.isna(series.std()):
        return f"{series.median():.3f}"
    return f"{series.median():.3f}" + " ${}_{\pm" + f"{series.std():.3f}" + "}$"


# 1. Define your specific order
projection_order = [
    "cca", "svd", "dct", "random", 
    # "cca-1/2_svd",  "dct-1/2_svd", "random-1/2_svd"
]

# 2. Convert the column to a Categorical type with that order
d['projection'] = pd.Categorical(d['projection'], categories=projection_order, ordered=True)

s = d.groupby(['dataset', 'lora_r', 'posterior', "projection"]).agg({"Accuracy": stat, "ECE": stat, "NLL": stat})

s
```
